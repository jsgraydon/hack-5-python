```python
from io import StringIO
import numpy as np
import pandas as pd
import requests
from scipy.stats import ttest_ind
```

```python
response = requests.get("http://eaton-lab.org/data/iris-data-dirty.csv")

csv_data = StringIO(response.text)

df = (pd.read_csv(csv_data, names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species_ID"])
        .dropna()
        .replace('Iris-setsa', 'Iris-setosa')
        .replace('Iris-versicolour', 'Iris-versicolor'))
```

```python
df.describe()
```

```python
for species,group in df.groupby("species_ID"):
    print(f"Summary statistics for {species}:")
    print("")
    print(group.describe())
    print("_" * 59 + "\n")
```

## Conclusions

A brief review of the summary statistics above suggests a few facts about the iris species included:
- *I. virginica* has the largest mean sepal and petal sizes
- *I. setosa* has the smallest mean sepal and petal sizes
- *I. setosa* has the smallest mean petal to sepal size ratios
- *I. setosa* tends to have the greatest relative variability in petal size
- All species have generally consistent petal length and width ratios
